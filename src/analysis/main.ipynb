```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
results_csv = '../../results/res_03.csv'
df = pd.read_csv(results_csv)

all_flags = df['flag'].unique()
flags = np.delete(all_flags, np.argwhere(all_flags == '-O0'))
progs = df['program'].unique()

print(flags)
print(progs)

df['total_energy'] = np.add(df['pkg1_energy'], df['pkg2_energy'])
```

```python
for flag in flags:
    flag_df = df[(df['flag'] == flag) | (df['flag'] == '-O0')]

    prog_ranges = {}

    for prog in progs:
        data = df[df['program'] == prog]['total_energy']
        prog_ranges[prog] = [np.min(data), np.max(data)]

    # divide programs into charts according to the ranges

    sns.boxplot(data = flag_df, y='program', x='total_energy', hue = 'flag')
    plt.show()
```
